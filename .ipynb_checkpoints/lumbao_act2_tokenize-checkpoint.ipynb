## Text Normalization

### 1. Tokenizing data using Penn Tree Bank Tokenization

````python
import nltk
nltk.download('gutenberg')
````

````python
hamletData=nltk.corpus.gutenberg.raw('shakespeare-hamlet.txt')
print(hamletData)
````

````python
ptb_tokenizer = nltk.tokenize.treebank.TreebankWordTokenizer()
PTB_TOKENS = ptb_tokenizer.tokenize(hamletData)
````

````python
print(PTB_TOKENS)
````

### 2. Apply case folding on the tokens (```lower case```).


````python
CF_PTB_TOKENS = [w.lower() for w in PTB_TOKENS]
print(CF_PTB_TOKENS)
````

### 2. Apply case folding on the tokens (```UPPER CASE```).

````python
CF_PTB_TOKENS = [w.upper() for w in PTB_TOKENS]
print(CF_PTB_TOKENS)
````

### 3. Lemmatize the case folded tokens using Word Net Lemmatizer

````python
nltk.download('wordnet')

lemmatizer = nltk.WordNetLemmatizer()
WNL_PTB_TOKENS = [lemmatizer.lemmatize(w) for w in PTB_TOKENS]
print(WNL_PTB_TOKENS)
````

### 4. From the lemmatized tokens, calculate the following metrics
```
a. Word type (Number of unique tokens)
b. Word Token (Number of tokens) ```

#### A. Number of unique tokens

````python
uniqueString = " ".join(map(str,WNL_PTB_TOKENS))

words_string = uniqueString.split(" ")

unique_words = []

for word in words_string:
    if word not in unique_words:
        unique_words.append(word)
    else:
        continue
        
# print(len(unique_words))
print("\nNumber of unique tokens: " + str(len(unique_words)))
````

#### B. Number of tokens

````python
total_tokens = WNL_PTB_TOKENS
print("\nNumber of tokens: " + str(len(total_tokens)))
````
